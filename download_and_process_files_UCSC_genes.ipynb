## Iteratively create 1000 random shuffles and intersections and consolidate them

```python
import os

from tqdm import tqdm

# --- Configuration ---
NUM_PERMUTATIONS = 1000
GENOME_FILE = "chm13.genome"
INPUT_REPEATS = "repeats_all.bed"
TSS_FILE = "BE2C.H3K9me3.chm13v2.0.mapped_on_TSS.bedGraph" # Target neighborhood
OUTPUT_DIR = "permutation_results"
N_JOBS = os.cpu_count() or 4 # Use available CPU cores for parallelism

# --- Logging and File Setup ---
bash_script_path = "run_permutation_test.sh"

print(f"INFO: Generating parallel bash script: {bash_script_path}")
print(f"INFO: Targeting {NUM_PERMUTATIONS} permutations using {N_JOBS} cores.")

with open(bash_script_path, 'w') as script:
    
    # --- Bash Script Header ---
    script.write("#!/bin/bash\n")
    script.write("# Script generated by Python to run bedtools permutation tests in parallel.\n")
    script.write("set -eo pipefail\n\n")

    # --- Variables and Environment Setup ---
    script.write(f"N_JOBS={N_JOBS}\n")
    script.write(f"OUTPUT_DIR={OUTPUT_DIR}\n")
    script.write(f"GENOME_FILE={GENOME_FILE}\n")
    script.write(f"INPUT_REPEATS={INPUT_REPEATS}\n")
    script.write(f"TSS_FILE={TSS_FILE}\n")
    
    script.write("\nmkdir -p ${OUTPUT_DIR}\n")
    
    # Define color logging for visibility
    script.write('GREEN="\\033[0;32m"\n')
    script.write('NC="\\033[0m"\n')
    script.write('echo -e "${GREEN}Starting permutation test...${NC}"\n\n')

    # --- Core Processing Function ---
    script.write("# Function to run one permutation (takes seed number as argument)\n")
    script.write("process_permutation() {\n")
    script.write("    SEED=$1\n")
    script.write("    TEMP_SHUFFLE=${OUTPUT_DIR}/repeats_all_random_${SEED}.bed\n")
    script.write("    OUTPUT_INTERSECT=${OUTPUT_DIR}/repeats_intersected_with_TSS_random_${SEED}.bed\n")
    
    # Shuffle step
    script.write("    bedtools shuffle -i ${INPUT_REPEATS} -g ${GENOME_FILE} -seed ${SEED} > ${TEMP_SHUFFLE}\n")
    
    # Intersect step
    script.write("    bedtools intersect -a ${TEMP_SHUFFLE} -b ${TSS_FILE} -wa > ${OUTPUT_INTERSECT}\n")
    
    # Cleanup step
    script.write("    rm ${TEMP_SHUFFLE}\n")
    
    script.write('    echo -e "${GREEN}  [DONE] Permutation ${SEED} processed.${NC}"\n')
    script.write("}\n")
    
    script.write("export -f process_permutation\n\n")

    # --- Parallel Execution ---
    script.write("# Pipe 1 to 1000 seeds into xargs to run in parallel\n")
    script.write("seq 1 1 ${NUM_PERMUTATIONS} | xargs -n 1 -P ${N_JOBS} bash -c 'process_permutation \"$0\"'\n\n")
    
    script.write('echo -e "${GREEN}All 1000 permutations finished.${NC}"\n')
    script.write("# The output files are saved in the ${OUTPUT_DIR} directory.\n")
```

### Combine the bed files into the single large consolidated file for efficiency

```python


# --- Configuration ---
PERMUTATION_DIR = "./epigenomic_files/permutation_results"
# Loop range is 1 to 500 (500 permutations total)
NUM_PERMUTATIONS = 500 
# The final file where all data will be stored
FINAL_CONSOLIDATED_FILE = os.path.join(PERMUTATION_DIR, "consolidated_random_data.csv")

# Column names based on your intersection output (after the first 3 columns are cut)
COL_NAMES = ["score", "subfamily_name", "family_name", "class_name"]

# --- 1. Efficient Consolidation and Tagging ---

def consolidate_files_with_id_tagging():
    """
    Reads each variable-length permutation file, tags each row with 
    its permutation ID, and writes it to a single large CSV file.
    """
    if os.path.exists(FINAL_CONSOLIDATED_FILE):
        os.remove(FINAL_CONSOLIDATED_FILE)
    
    tqdm.write("INFO: Starting file consolidation and permutation tagging...")
    
    # Use 'w' mode to create the file and add the header row
    with open(FINAL_CONSOLIDATED_FILE, 'w') as outfile:
        # Write the header for the final CSV
        header = "permutation_id\t" + "\t".join(COL_NAMES) + "\n"
        outfile.write(header)
        
        # Iterate over all permutation IDs
        for i in tqdm(range(1, NUM_PERMUTATIONS + 1), desc="Consolidating Permutations"):
            input_file = os.path.join(PERMUTATION_DIR, f"repeats_intersected_with_TSS_random_{i}.bed")
            
            # Use Python's built-in file handling for speed
            try:
                with open(input_file, 'r') as infile:
                    # Write the content line-by-line
                    for line in infile:
                        # Prepend the permutation ID (i) and a tab
                        outfile.write(f"{i}\t{line}")
                        
            except FileNotFoundError:
                tqdm.write(f"WARNING: File not found for permutation {i}. Skipping.")
                continue
            
    tqdm.write(f"SUCCESS: All available files consolidated into {FINAL_CONSOLIDATED_FILE}")
```
